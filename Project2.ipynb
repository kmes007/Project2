```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('sqlite:///db/Performance.sqlite')
```

```python
scores = pd.read_excel("db/Performance.xlsx")
```

```python
scores['Mean Scale Score'] = pd.to_numeric(scores['Mean Scale Score'], errors='coerce')#.astype(float, errors='coerce').dtype
```

```python
scores.head()
```

```python
mean_by_district = scores.groupby('State/District/School')['Mean Scale Score'].mean()
```

```python
mean_by_district.to_json(orient='split')
```

```python
scores.to_sql('scores3', con=engine)
```

```python
scores.head(100)
```

```python
engine.execute("SELECT * FROM scores3").fetchall()
```

```python
turnover_rate = pd.read_excel("db/PersonnelTurnoverRate_5Yr.xlsx")
```

```python
turnover_rate.to_sql("turnover_rate3", con=engine)
```

```python
engine.execute("SELECT * FROM turnover_rate3").fetchall()
```

```python
turnover_rate["District Name"].unique()
```

```python
districts =  ["0020","0030","0180", "0480","1040", "0070","0120", "0130", 
              "0140","0640","0880", "0900","1010","1420"]
district_final = turnover_rate[turnover_rate["District Code"].isin(districts)]


 
```

```python
district_final["District Name"].unique()
```

```python
district_final.head(100)
```

```python
mean = district_final[['2014 Turnover Rate', '2015 Turnover Rate','2016 Turnover Rate','2017 Turnover Rate', '2018 Turnover Rate' ]].mean(axis=1)
district_final['Mean'] = mean
district_final.head(20)
```

```python
import matplotlib.pyplot as plt
by_job = district_final.groupby("Job Classification")
by_job.head()
```

```python
%matplotlib inline
ax = plt.gca()
by_job.plot(kind="line", x="Job Classification", y="Mean",ax=ax)
labels = by_job["Job Classification"].unique()
plt.legend(labels, loc="lower right", fontsize = "x-small")
plt.show()
```

```python
teacher_rows = ["Teachers"]

teacher_retention = district_final[district_final["Job Classification"].isin(teacher_rows)] 
```

```python
teacher_retention.head(20)
```

```python
other_rows = ["Other Support", "Teachers"]
other_retention = district_final[district_final["Job Classification"].isin(other_rows)]
```

```python
ax = plt.gca()
other_retention.plot(kind="bar", x="Job Classification", y="Mean", ax=ax)
plt.legend(loc="lower right", fontsize = "x-small")
plt.show()
```

```python
teacher_retention.head()
```

```python
teacher_retention.to_sql("retention", con=engine)
```

```python
engine.execute("SELECT * FROM retention").fetchall()
```

```python
other_retention.to_sql("other", con=engine)
```

```python
engine.execute("SELECT * FROM other").fetchall()
```
